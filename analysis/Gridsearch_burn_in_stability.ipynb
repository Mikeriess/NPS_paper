# Convergence criteria 1: Variation
- <b>Criteria 1</b> Stabilisation achieved in burn-in period (convergence criterion met, e.g. variance over last eight weeks < c, or some SPC stability index)

```python
#https://falltechnicalconference.org/wp-content/uploads/2017_Presentations/4B-2-Szarka-Stability-Assessment-with-the-Stability-Index.pdf
```

```python
def variance(data):
    # Number of observations
    n = len(data)
    # Mean of the data
    mean = sum(data) / n
    # Square deviations
    deviations = [(x - mean) ** 2 for x in data]
    # Variance
    variance = sum(deviations) / n
    return variance
```

```python
def sliding_window(data, x="FCFS_1", window_size=60):
    """
    define performance metrics
    """
    def variance(data):
        # Number of observations
        n = len(data)
        
        # Mean of the data
        mean = sum(data) / n
        
        # Square deviations
        deviations = [(x - mean) ** 2 for x in data]
        
        # Variance
        variance = sum(deviations) / n
        return variance
    
    """
    perform the sliding window
    """
    #placeholder
    process_performance = []
    
    #go over the period
    for day in range(0,len(data)):
        
        #if there is enough history for the window
        if day >= window_size-1:
            
            #subset the data for current sliding window
            window = data.loc[day-(window_size-1):day]
            
            #controlchart: normal
            #std=np.std(window[x])
            #UCL=np.mean(window[x])+(std*3)
            #LCL=np.mean(window[x])-(std*3)
            
            #controlchart: poisson
            c = np.sum(window[x])/len(window[x])
            
            UCL = c + (3*np.sqrt(c))
            LCL = c - (3*np.sqrt(c))
            
            if LCL <= 0:
                LCL = 0
            
            x_val = window.loc[day,x]
            
            #crate a dataframe with results for the day
            res_i = pd.DataFrame({"day":day,
                                    "x":x_val,
                                    "variance":variance(window[x]),
                                    "sd":np.std(window[x]),
                                    "c":c,
                                    "UCL":UCL,
                                    "LCL":LCL,
                                    #"control":(x <),
                                    "mean":np.mean(window[x]),
                                    "min":np.min(window[x]),
                                    "max":np.max(window[x]),
                                    "median":np.median(window[x])},
                                                  index=[day])

            #append to history
            process_performance.append(res_i)
            
    process_performance = pd.concat(process_performance)
    return process_performance
```

```python
process_performance = sliding_window(results, x="FCFS_21", window_size=90)
```
