```python
import numpy as np
```

```python
mu = 1.5
sigma = 1
```

```python
s = np.random.normal(mu, sigma, 1000)
s
```

```python
def sigmoid_array(x):
    return 1 / (1 + np.exp(-x))
```

```python
s = sigmoid_array(s)
s
```

```python
s = np.floor(s*11)
```

```python
s
```

```python
import pandas as pd

df = pd.DataFrame({"NPS":s})

df.NPS.hist(bins=11)
```

```python
df.NPS.describe()
```
