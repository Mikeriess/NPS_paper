# Averages per experimental condition

```python
runs = []
for i in list(range(0,int(np.max(experiments.RUN)))):
    
    # load a single run
    df = pd.read_csv("../results/"+str(i)+"/"+str(i)+"_timeseries.csv")
    
    # mark it with specs
    specs = experiments.loc[i].to_dict()
    
    df["F_priority_scheme"] = specs["F_priority_scheme"]
    df["F_number_of_agents"] = specs["F_number_of_agents"]
    df["F_hard_ceiling"] = specs["F_hard_ceiling"]
    
    df["repetition"] = specs["repetition"]
    df["RUN"] = specs["RUN"]
    
    """
    Calculate needed metrics:
    % cases closed (cases arrived / cases closed)
    Average capacity utilization (avg. number of busy agents / number of agents)

    """
    
    df["percent_closed"] = df["n_cases_caseDB"] / df["n_cases_closed_total"]
    df["capacity_utilization"] = df["n_agents"] / (df["n_busy_agents"]+.001)
    
    # add it to a list
    runs.append(df)
    
all_runs = pd.concat(runs)

all_runs["F_priority_scheme"] = all_runs.F_priority_scheme.astype("category")
all_runs["F_number_of_agents"] = all_runs.F_number_of_agents.astype("category")
all_runs["F_hard_ceiling"] = all_runs.F_hard_ceiling.astype("category")
```

```python
all_runs.F_priority_scheme.unique()
```

```python
agg_runs = []

for discipline in all_runs.F_priority_scheme.unique():
    
    for n_agents in all_runs.F_number_of_agents.unique():
        
        for SLA in all_runs.F_hard_ceiling.unique():
            
            subset = all_runs.loc[all_runs.F_priority_scheme == discipline]
            
            subset = subset.loc[subset.F_number_of_agents == n_agents]
            subset = subset.loc[subset.F_hard_ceiling == SLA]
            
            for dayno in subset.day.unique():
                
                daydata = subset.loc[subset.day == dayno]
                
                """
                Metrics
                """

                subdf = pd.DataFrame({
                                      "day":dayno,
                                      "discipline":discipline,
                                      "n_agents":n_agents,
                                      "SLA":SLA,
                                      "avg_avg_queue_waitingtime":np.mean(daydata["avg_current_queue_waitingtime"]),
                                      "avg_queue_length":np.mean(daydata["n_cases_waiting_in_queue"]),
                                      "avg_capacity_utilization":np.mean(daydata["n_agents"] / daydata["n_busy_agents"]),
                                         },
                                     index=[dayno])

                agg_runs.append(subdf)

agg_runs = pd.concat(agg_runs)
```

```python
# change inf to 0
#agg_runs.replace([np.inf, -np.inf], 0, inplace=True)

# fix index issues
agg_runs.index = list(range(0,len(agg_runs)))

agg_runs
```
